```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
finalRatings,finalTitles,finalReviews,finalProfile=[],[],[],[]
```

```python
stars=['one','two','three','four','five']
```

```python
for star in stars:
  ratingTitle=[]
  url=f"https://www.amazon.in/Echo-Dot-5th-Gen-Alexa-smart-speaker/product-reviews/B09B8XJDW5/filterByStar=one_star/ref=cm_cr_arp_d_viewopt_sr?filterByStar={star}_star"
  response=requests.get(url)
  if response.status_code==200:
    print(url)
    print(response)
    rating=[]
    title=[]
    soup=bs(response.content,"html.parser")
    soup.prettify()
    tr=soup.find_all('a',class_="review-title")
    for x in tr:
      ratingTitle.append(x.text.split('\n'))
    for x in range(len(ratingTitle)):
      rating.append(float(ratingTitle[x][0].split(' ')[0]))
      title.append(ratingTitle[x][1])
    review=soup.find_all('div',class_="a-row a-spacing-small review-data")
    for x in range(len(review)):
      review[x]=review[x].text.replace('\n','')
    profileName=soup.find_all('span',class_="a-profile-name")
    for x in range(len(profileName)):
      profileName[x]=profileName[x].text
    profileName.pop(0)
    profileName.pop(0)
    if(star!='one'):
      profileName=list(dict.fromkeys(profileName))
    finalRatings.extend(rating)
    finalTitles.extend(title)
    finalReviews.extend(review)
    finalProfile.extend(profileName)
```

```python
#create a dataframe of obtained data
import pandas as pd
df=pd.DataFrame({'Rating':finalRatings,'Title':finalTitles,'Review':finalReviews,'Profile':finalProfile})
```

## Preprocessing the text for sentiment analysis

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
def preprocess_text(text):
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove stopwords
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text
```

```python
df['cleaned_review'] = df['Review'].apply(preprocess_text)
```

## Perform sentiment analysis using textblob

```python
from textblob import TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    return 'positive' if analysis.sentiment.polarity > 0 else 'negative'
```

```python
df['sentiment'] = df['cleaned_review'].apply(get_sentiment)
```
