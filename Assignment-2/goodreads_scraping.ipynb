Scraping goodreads.com to get reviews of the book series "The Hitchhiker's Guide to The Galaxy" and then performing sentiment analysis to classify the review into positive and negative

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
page=requests.get('https://www.goodreads.com/book/show/372299/reviews?reviewFilters={%22workId%22:%22kca://work/amzn1.gr.work.v1.TzH3ypxiQeGk1IVy6GHmcA%22,%22after%22:%22OTEwLDE1MzM0NDkxMTE5MTQ%22}')
page
```

```python
soup=bs(page.content,'html.parser')
```

```python
reviews=soup.find_all('span',class_='Formatted')
reviews
```

```python
review_body=[]
for review in reviews:
    review_body.append(review.text.strip())
review_body
```

```python
ratings = soup.find_all('span', class_='RatingStars')
rating_values = []
for rating in ratings:
    aria_label = rating.get('aria-label')
    if aria_label:
        rating_value = float(aria_label.split(' ')[1])
        rating_values.append(rating_value)
```

# Sentiment analysis using textblob

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
def preprocess_text(text):
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove stopwords
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text
```

```python
df['cleaned_reviews'] = df['Review'].apply(preprocess_text)
```

```python
#perform sentiment analysis using textblob
from textblob import TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    Polarity=analysis.sentiment.polarity
    if Polarity > 0:
        return 'positive'
    elif Polarity < 0:
        return 'negative'
    else:
        return 'neutral'
def get_polarity(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
```

```python
df['Polarity'] = df['cleaned_reviews'].apply(get_polarity)
df['Sentiment'] = df['cleaned_reviews'].apply(get_sentiment)
```
